# **Step 1: Read Text Files**

```python
# Example: Read a text file from your local machine or Google Drive
resume_text = """
Ouday BOUASKER
Email: bouasker.ouday@gmail.com
Experienced Software Engineer with knowledge in Python, AI, and Automation.
Skilled in NLP, Machine Learning, and Robotics.
"""

print(resume_text)
```

# **Step 2: Clean the Text**

NLP works better when text is standardized.

```python
# Convert text to lowercase
clean_text = resume_text.lower()

# Remove punctuation
import string
clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))

print(clean_text)
```

**Why this Phase !!**

*   Standardization prevents “Python” and “python” from being treated as two separate words.

*   Removes clutter that doesn’t carry meaning.



# **Step 3: Tokenize the Text**

Break the text into words.

```python
# Simple tokenization
words = clean_text.split()  # Splits by spaces
print(words)
```

Optional Extension: Use nltk for more advanced tokenization:

```python
# Install NLTK if needed
!pip install nltk

import nltk
nltk.download('punkt_tab') # Changed 'punkt' to 'punkt_tab' as indicated by the error
nltk.download('punkt') # Also ensure 'punkt' is downloaded, as it is generally useful.
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(clean_text)
print(tokens)
```

# **Step 4: Count Word Occurrences**

```python
from collections import Counter

word_counts = Counter(words)
print(word_counts)
```
