**üéØ Objective**

Automatically extract technical skills from a resume text using Python.
You‚Äôll use a predefined list of skills as a ‚Äúdictionary‚Äù and match them in the text.

This is the first step toward making your AI resume analyzer intelligent.

# **Step 1: Define Your Skill List**

```python
# Example skill dictionary
skills = [
    "python", "c++", "embedded", "automation", "nlp", "machine learning",
    "robotics", "do-178c", "iso 26262", "testing", "validation"
]
```

# **Step 2: Preprocess Resume Text**

```python
import string

resume_text = """
Ouday BOUASKER
Email: bouasker.ouday@gmail.com
Experienced Embedded Software Validation Engineer with knowledge in Python, AI,
Automation, Robotics, and Testing. Skilled in DO-178C, ISO 26262, and embedded systems.
"""

# Preprocess
clean_text = resume_text.lower()
print(clean_text)
clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)
tokens = clean_text.split()
print(tokens)
```

# **Step 3: Match Skills in Text**

```python
found_skills = [skill for skill in skills if skill in tokens]
print("Skills found in resume:", found_skills)
```

# **Step 4: Count Skill Occurrences (Optional)**

```python
skill_counts = {skill: tokens.count(skill) for skill in found_skills}
print("Skill counts:", skill_counts)
```

# **Step 5: Display Results Nicely**

```python
print("\nExtracted Skills and Frequencies:")
for skill, count in skill_counts.items():
    print(f"{skill}: {count}")
```
