Build a Python notebook that reads resume text, cleans it, tokenizes it, removes stopwords, and prints the top 10 most frequent words.

# **1Ô∏è‚É£ Import Libraries**

```python
# For NLTK tokenization and stopwords
import nltk
nltk.download('punkt')        # for tokenization
nltk.download('punkt_tab')    # for tokenization
nltk.download('stopwords')    # for common English words
```

```python
import string
from collections import Counter

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

# **2Ô∏è‚É£ Load Resume Text**

```python
#open("file.txt").read()

resume_text = """
Ouday BOUASKER
Email: bouasker.ouday@gmail.com
Experienced Embedded Software Validation Engineer with knowledge in Python, AI,
Automation, Robotics, and Testing. Skilled in DO-178C, ISO 26262, and embedded systems.
"""
print(resume_text)
```

# **3Ô∏è‚É£ Clean the Text**

```python
# Convert to lowercase
clean_text = resume_text.lower()
print(clean_text)
```

```python
# Remove punctuation
clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)
```

# **4Ô∏è‚É£ Tokenize the Text**

```python
tokens = word_tokenize(clean_text)
tokens
```

# **5Ô∏è‚É£ Remove Stopwords**

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_tokens
```

# **6Ô∏è‚É£ Count Word Frequencies**

```python
word_freq = Counter(filtered_tokens)
word_freq
```

# **7Ô∏è‚É£ Display Top 10 Most Frequent Words**

```python
top_10 = word_freq.most_common(10)
print("Top 10 Most Frequent Words in R√©sum√©:")
for word, freq in top_10:
    print(f"{word}: {freq}")
```
